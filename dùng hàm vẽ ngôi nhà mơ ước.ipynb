```python
import turtle

tl = turtle.Turtle()
tl.shape("turtle")
tl.pensize(1)
tl.hideturtle()
tl.pencolor("Maroon")
tl.speed(9)
screen =tl.getscreen()
screen.bgcolor("lightskyblue")

def Rectangle_draw(w,h):
    '''vẽ hình chữ nhật'''
    for i in range(2):
        tl.forward(w)
        tl.right(90)
        tl.forward(h)
        tl.right(90)
        
def triangle_draw(t,f):
    '''vẽ hình tam giác với cạnh t và góc f'''
    for i in range(3):
        tl.fd(t)
        tl.lt(f)
        tl.fd(t)

tl.fillcolor("yellow")
tl.begin_fill()
tl.pu()
tl.bk(120)
tl.pendown()
Rectangle_draw(200,100)
tl.end_fill()
tl.fillcolor("red")
tl.begin_fill()
tl.lt(45)
tl.fd(90)
tl.rt(90)
tl.fd(90)
tl.end_fill()

tl.fillcolor("brown")
tl.begin_fill()
tl.bk(30)
tl.lt(135)
tl.fd(45)
tl.lt(90)
tl.fd(20)
tl.lt(90)
tl.fd(30)
tl.end_fill()

tl.pu()
tl.goto(-40,-40)
tl.pendown()
tl.fillcolor("Olive")
tl.begin_fill()
tl.lt(90)
Rectangle_draw(40,60)
tl.end_fill()

for i in range(1,4):
    tl.pu()
    tl.goto(200,-50*i)
    tl.pendown()
    tl.fillcolor("Green")
    tl.begin_fill()
    triangle_draw(40,120)
    tl.end_fill()
tl.bk(5)
tl.fillcolor("brown")
tl.begin_fill()
Rectangle_draw(10,40)
tl.end_fill()

for i in range(1,4):
    tl.pu()
    tl.goto(-200,-50*i)
    tl.pendown()
    tl.fillcolor("Green")
    tl.begin_fill()
    triangle_draw(40,120)
    tl.end_fill()
tl.bk(5)
tl.fillcolor("brown")
tl.begin_fill()
Rectangle_draw(10,40)
tl.end_fill()
      
turtle.done()
```
