```python
# vẽ hình vuông
import turtle

t = turtle.Turtle()
t.hideturtle()
t.pencolor('black')
screen =t.getscreen()
screen.bgcolor("lightskyblue")

def Rectangle_draw(w,h):
    '''vẽ hình chữ nhật'''
    for i in range(2):
        t.forward(w)
        t.right(90)
        t.forward(h)
        t.right(90)
   
t.fillcolor('Silver')
t.begin_fill()
Rectangle_draw(130,150)
t.end_fill()

t.fillcolor('red')
t.begin_fill()
t.penup()
t.goto(35,-60)
t.pendown()
Rectangle_draw(60,90)
Rectangle_draw(30,90)
t.end_fill()

for i in range(1,3):
    t.pu()
    t.goto(-100*i,0)
    t.pd()
    t.fillcolor('#FF66FF')
    t.begin_fill()
    Rectangle_draw(100,150)
    t.end_fill()
    t.pu()
    t.goto(-100*i+35,-60)
    t.pd()
    t.fillcolor('white')
    t.begin_fill()
    Rectangle_draw(30,50)
    t.end_fill()
for i in range(1,3):
    t.pu()
    t.goto(-100*i,150)
    t.pd()
    t.fillcolor('#FF66FF')
    t.begin_fill()
    Rectangle_draw(100,150)
    t.end_fill()
    t.pu()
    t.goto(-100*i+35,90)
    t.pd()
    t.fillcolor('white')
    t.begin_fill()
    Rectangle_draw(30,50)
    t.end_fill()




turtle.done()
```
