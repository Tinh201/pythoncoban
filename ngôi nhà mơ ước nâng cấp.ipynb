```python
import turtle

tl = turtle.Turtle()
tl.shape("turtle")
tl.pensize(1)
tl.hideturtle()
tl.pencolor("Maroon")
tl.speed(9)
screen =tl.getscreen()
screen.bgcolor("lightskyblue")

#a = 50
tl.fillcolor("yellow")
tl.begin_fill()
tl.pu()
tl.bk(120)
tl.pendown()
tl.fd(200)
tl.rt(90)
tl.fd(100)
tl.rt(90)
tl.fd(200)
tl.rt(90)
tl.fd(100)
tl.end_fill()

tl.fillcolor("red")
tl.begin_fill()
tl.rt(45)
tl.fd(90)
tl.rt(90)
tl.fd(90)
tl.end_fill()

tl.fillcolor("brown")
tl.begin_fill()
tl.bk(30)
tl.lt(135)
tl.fd(45)
tl.lt(90)
tl.fd(20)
tl.lt(90)
tl.fd(30)
tl.end_fill()

tl.pu()
tl.goto(0,-100)
tl.pendown()
tl.fillcolor("Olive")
tl.begin_fill()
tl.bk(60)
tl.rt(90)
tl.fd(40)
tl.lt(90)
tl.fd(60)
tl.lt(90)
tl.end_fill()

tl.pu()
tl.goto(200,50)
tl.pendown()
tl.fillcolor("Green")
tl.begin_fill()
for i in range(3):
    tl.fd(40)
    tl.lt(120)
    tl.fd(40)
tl.end_fill()   
tl.pu()
tl.goto(200,0)
tl.pendown()
tl.fillcolor("Green")
tl.begin_fill()
for i in range(3):
    tl.fd(40)
    tl.lt(120)
    tl.fd(40)
tl.end_fill()
tl.pu()
tl.goto(200,-50)
tl.pendown()
tl.fillcolor("Green")
tl.begin_fill()
for i in range(3):
    tl.fd(40)
    tl.lt(120)
    tl.fd(40)
tl.end_fill()
tl.rt(90)
tl.pensize(20)
tl.pencolor("brown")
tl.fd(40)

tl.pensize(1)
tl.lt(90)
tl.pu()
tl.goto(-200,-50)
tl.pendown()
tl.fillcolor("Green")
tl.begin_fill()
for i in range(3):
    tl.fd(20)
    tl.lt(120)
    tl.fd(20)
tl.end_fill()   
tl.pu()
tl.goto(-200,-100)
tl.pendown()
tl.fillcolor("Green")
tl.begin_fill()
for i in range(3):
    tl.fd(30)
    tl.lt(120)
    tl.fd(30)
tl.end_fill()
tl.pu()
tl.goto(-200,-150)
tl.pendown()
tl.fillcolor("Green")
tl.begin_fill()
for i in range(3):
    tl.fd(40)
    tl.lt(120)
    tl.fd(40)
tl.end_fill()
tl.rt(90)
tl.pensize(20)
tl.pencolor("brown")
tl.fd(40)
turtle.done()
```
