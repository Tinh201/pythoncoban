```python
import turtle

tl = turtle.Turtle()
tl.shape("turtle")
tl.pensize(1)
tl.hideturtle()
tl.pencolor("black")
tl.speed(9)
screen =tl.getscreen()
screen.bgcolor("lightskyblue")

a = 30
tl.fillcolor("yellow")
tl.begin_fill()
tl.rt(90)
tl.fd(a*5)
tl.rt(60)
tl.fd(a*5)
tl.rt(120)
tl.fd(a*5)
tl.rt(60)
tl.fd(a*5)
tl.end_fill()

tl.fillcolor("red")
tl.begin_fill()
tl.lt(90)
tl.fd(a*4)
tl.lt(90)
tl.fd(a*5)
tl.lt(90)
tl.fd(a*4)
tl.end_fill()


tl.fillcolor("Purple")
tl.begin_fill()
tl.rt(120)
tl.fd(a*4)
tl.rt(120)
tl.fd(a*4)
tl.end_fill()

tl.pu()
tl.bk(120)
tl.pendown()
tl.fillcolor("blueviolet")
tl.begin_fill()
tl.rt(150)
tl.fd(a*5)
tl.lt(90)
tl.fd(a*4)
tl.lt(90)
tl.fd(a*5)
tl.end_fill()

tl.pu()
tl.bk(150)
tl.pendown()
tl.fillcolor("brown")
tl.begin_fill()
tl.rt(90)
tl.bk(45)
tl.lt(90)
tl.fd(60)
tl.lt(90)
tl.fd(30)
tl.lt(90)
tl.fd(60)
tl.end_fill()

tl.pu()
tl.goto(-40,-90)
tl.pendown()
tl.fillcolor("brown")
tl.begin_fill()
tl.fd(30)
tl.rt(60)
tl.fd(30)
tl.rt(120)
tl.fd(30)
tl.rt(60)
tl.fd(30)
tl.end_fill()

tl.pu()
tl.goto(120,150)
tl.pendown()
tl.fillcolor("yellow")
tl.begin_fill()
tl.circle(50)
tl.end_fill()

tl.pu()
tl.rt(30)
tl.goto(200,-50)
tl.pendown()
tl.fillcolor("Green")
tl.begin_fill()
for i in range(3):
    tl.fd(60)
    tl.lt(120)
    tl.fd(60)
tl.end_fill()   
tl.pu()
tl.goto(200,-120)
tl.pendown()
tl.fillcolor("Green")
tl.begin_fill()
for i in range(3):
    tl.fd(60)
    tl.lt(120)
    tl.fd(60)
tl.end_fill()
tl.pu()
tl.goto(200,-200)
tl.pendown()
tl.fillcolor("Green")
tl.begin_fill()
for i in range(3):
    tl.fd(60)
    tl.lt(120)
    tl.fd(60)
tl.end_fill()
tl.rt(90)
tl.pensize(30)
tl.pencolor("brown")
tl.fd(50)

turtle.done()
```
